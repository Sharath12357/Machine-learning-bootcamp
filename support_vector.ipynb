```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000,n_classes=2,n_features=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
support=SVC(kernel='rbf')
```

```python
support.fit(X_train,y_train)
```

```python
y_predict=support.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
martix=confusion_matrix(y_test,y_predict)
```

```python
print(martix)
```

```python
report=classification_report(y_test,y_predict)
```

```python
print(report)
```

```python
score=accuracy_score(y_test,y_predict)
print(score)
```

```python
poloynomial=SVC(kernel='poly')
```

```python
poloynomial.fit(X_train,y_train)
```

```python
y_predct1=poloynomial.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
confusion=confusion_matrix(y_test,y_predct1)
classification=classification_report(y_test,y_predct1)
print(confusion)
print(classification)
```

```python
score=accuracy_score(y_test,y_predct1)
print(score)
```

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
y_prdct2=sigmoid.predict(X_test)
confusion=confusion_matrix(y_test,y_prdct2)
classification=classification_report(y_test,y_prdct2)
print(confusion)
print(classification)
score=accuracy_score(y_test,y_prdct2)
print(score)
```

```python
#Hyperparametrt Tuning with SVC
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1,0.01,0.001,0.0001],
            'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_prd=grid.predict(X_test)
```

```python
confusion=confusion_matrix(y_test,y_prd)
classification=classification_report(y_test,y_prd)
print(confusion)
print(classification)
```

```python
score=accuracy_score(y_test,y_prd)
print(score)
```
