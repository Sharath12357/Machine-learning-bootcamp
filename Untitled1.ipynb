```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Get data for Reliance Industries
reliance = yf.Ticker("RELIANCE.NS")

# Fetch historical price data (last 1 month)
data = reliance.history(period="1mo")
print(data.head())

# Get basic info
print(reliance.info['sector'])
data['7_day_MA'] = data['Close'].rolling(window=7).mean()

# Step 4: Analyze volume trends
high_volume_days = data[data['Volume'] > data['Volume'].mean()]
print("\nHigh volume trading days:")
print(high_volume_days[['Close', 'Volume']])

# Step 5: Plot the stock Close price and Moving Average
plt.figure(figsize=(12,6))
plt.plot(data.index, data['Close'], label='Close Price', color='blue')
plt.plot(data.index, data['7_day_MA'], label='7-Day MA', color='orange')
plt.title("Reliance Industries Stock - Close Price & Moving Avg")
plt.xlabel("Date")
plt.ylabel("Price (INR)")
plt.legend()
plt.grid()
plt.show()
```
